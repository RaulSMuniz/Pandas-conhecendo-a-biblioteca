# Aula 2 - Análise exploratória dos Dados

<h1>Qual o valor médio por tipo de imóvel?</h1>

```python
# Método groupby -> groupby('coluna').metodo(exemplo: mean)
# media_valor_imoveis = data.groupby('Tipo').mean(numeric_only=True)
media_valor_imoveis = data.groupby('Tipo')['Valor'].mean()
media_valor_imoveis.round(2)
```

```python
media_valor_imoveis = data.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
media_valor_imoveis.round(2)
```

```python
media_valor_imoveis.plot(kind='barh', figsize=(14,10), color='red')
```

<h1>Remoção de imóveis comerciais</h1>

```python
# data.Parametro.unique() -> Mostra todos os tipos individuais da coluna Tipo
data.Tipo.unique()
```

```python
imoveis_comerciais = ['Loja Shopping/ Ct Comercial',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Sítio', 'Hotel',
                      'Box/Garagem',
                      'Conjunto Comercial/Sala', 'Terreno Padrão',
                      'Chácara', 'Indústria',
                      'Pousada/Chalé', 'Loteamento/Condomínio', 'Casa Comercial']
```

```python
# Para passar uma variável no método 'query', o @ é necessário
comerciais = data.query('@imoveis_comerciais in Tipo')
comerciais
# Faz um dataframe apenas com os imoveis comerciais e armazena em 'comerciais'
```

```python
# Agrupa o DataFrame apenas pelo parametro 'Tipo' com o parametro 'Valor', e faz a média disso
# Para que uma determinada variável não esteja inclusa no query, é necessário
# adicionar um 'not' antes do 'in'
residenciais = data.query('@imoveis_comerciais not in Tipo')
media_residenciais = residenciais.groupby('Tipo')['Valor'].mean().sort_values()
media_residenciais

media_residenciais.plot(kind='barh', figsize=(14, 10), color='purple')
```

<h1>Qual o percentual de cada tipo de imóvel na nossa base de dados?</h1>

```python
residenciais.Tipo.unique()
```

```python
residenciais.Tipo.value_counts(normalize=True)
```

```python
percentual_tipo_residenciais = residenciais.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')

percentual_tipo_residenciais.plot(kind='bar', figsize=(14, 10), color='green', xlabel='Tipos', ylabel='Percentual')
```

# Desafio 2 da Aula 3: tratamentos e filtragem da base de dados 'alunos'

Desafio 1: Verifique se a base de dados possui dados nulos e, caso tenha, realize o tratamento desses dados nulos da forma que achar mais coerente com a situação.

Desafio 2: Os alunos "Alice" e "Carlos", não fazem mais parte da turma. Sendo assim, remova-os da base de dados.

Desafio 3: Aplique um filtro que selecione apenas os alunos que foram aprovados.

Desafio 4: Salve o DataFrame que possui apenas os alunos aprovados em um arquivo csv chamado "alunos_aprovados.csv".

```python
url = "https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv"
df_alunos = pd.read_csv(url)
df_alunos
```

```python
# Desafio 1: Verifique se a base de dados possui dados nulos e, caso tenha, realize o tratamento desses dados nulos da forma que achar mais coerente com a situação.
# df_alunos.isnull().sum() # Total de 6 notas nulas

df_alunos.fillna(0, inplace=True) # As notas nulas serão alteradas diretamente para 0
df_alunos
```

```python
# Desafio 2: Os alunos "Alice" e "Carlos", não fazem mais parte da turma. Sendo assim, remova-os da base de dados.

# Armazena os indexes que contém 'Carlos' e 'Alice'
remocao_indexes = df_alunos.query("Nome == 'Carlos' | Nome == 'Alice'").index

# Remove as linhas que contém indexes que contém Carlos e Alice
df_alunos.drop(remocao_indexes, axis=0, inplace=True)
df_alunos
```

```python
# Desafio 3: Aplique um filtro que selecione apenas os alunos que foram aprovados.

# Filtra e armazena em uma variável todos os alunos que estão aprovados
alunos_aprovados = df_alunos.query('Aprovado == True')
alunos_aprovados
```

```python
# Desafio 4: Salve o DataFrame que possui apenas os alunos aprovados em um arquivo csv chamado "alunos_aprovados.csv".

# Salva o DataFrame em um arquivo csv sem index 'Unnamed' e com separação padrão de arquivo csv
alunos_aprovados.to_csv("alunos_aprovados.csv", index=False, sep=';')
```

# Aula 4: manipulando os dados

<h1>Criando colunas numéricas</h1>

```python
url = "https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv"
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados['Valor_por_ano'] = (dados['Valor_por_mes'] * 12) + dados['IPTU']
dados.head()
```

<h1>Colunas Categóricas</h1>

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

```python
dados.to_csv("dados_completos.csv", index=False, sep=';')
```
